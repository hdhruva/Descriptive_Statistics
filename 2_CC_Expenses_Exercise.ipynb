## Data of 30 customers on credit card usage in the order of thousand INR
### sex: male - 1, female -2
### shopping or banking (yes-1, no-2)

```python
import pandas as mypd
myData=mypd.read_csv("C:/Users/jmo4cob/ETI/Data/CC_Expenses_Exercise.csv")
myData
```

```python
cc=myData.Credit_Card_usage
cc.describe()
```

```python
gender=myData.Sex
cc.groupby(gender).mean()
```

```python
shopping =myData.Shopping
banking=myData.Banking
cc.groupby(shopping).mean()
```

```python
cc.groupby(banking).mean()
```

```python
import matplotlib.pyplot as myplot
myplot.hist(cc)
myplot.show()
```

```python
myData.boxplot(column='Credit_Card_usage',by='Sex')
myplot.show()
```
